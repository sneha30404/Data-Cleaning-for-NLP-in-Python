```python
import pandas as pd

data = pd.read_pickle('dtm.pkl')
data = data.transpose()
data.head(20)
```

```python
word_counts = {comedian: {"like": 0, "love": 0} for comedian in data.columns}
```

```python
for comedian in data.columns:
    if "like" in data.index:
        word_counts[comedian]["like"] = data.loc["like", comedian]
    if "love" in data.index:
        word_counts[comedian]["love"] = data.loc["love", comedian]
word_counts
```

```python
like = [word_counts[comedian]["like"] for comedian in data.columns]
love= [word_counts[comedian]["love"] for comedian in data.columns]
comedians = list(data.columns)

like
```

```python
love
```

```python
import matplotlib.pyplot as plt
```

```python
for i, comedian in enumerate(comedians):
    plt.annotate(comedian, (like[i], love[i]))
    plt.scatter(like, love,color='blue')

plt.xlabel("like")
plt.ylabel("love")
plt.title("Comparison of like vs. love")
plt.show()
```
